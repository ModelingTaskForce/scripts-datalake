```python
rnm_columns = {
    'codigoEstadoTeste1': 'EstadoTeste1',
    'codigoEstadoTeste2': 'EstadoTeste2',
    'codigoEstadoTeste3': 'EstadoTeste3',
    'codigoEstadoTeste4': 'EstadoTeste4',
    
    'codigoResultadoTeste1': 'ResultadoTeste1',
    'codigoResultadoTeste2': 'ResultadoTeste2',
    'codigoResultadoTeste3': 'ResultadoTeste3',
    'codigoResultadoTeste4': 'ResultadoTeste4',
    
    'codigoTipoTeste1': 'TipoTeste1',
    'codigoTipoTeste2': 'TipoTeste2',
    'codigoTipoTeste3': 'TipoTeste3',
    'codigoTipoTeste4': 'TipoTeste4',
}
```

```python
var_locals = [
    'estado','estadoIBGE', 'municipio', 'municipioIBGE'
]

var_tests = [
    'EstadoTeste1', 'EstadoTeste2', 'EstadoTeste3', 'EstadoTeste4'
]

var_type_tests = [
    'TipoTeste1', 'TipoTeste2', 'TipoTeste3', 'TipoTeste4'
]

var_results = [
    'ResultadoTeste1', 'ResultadoTeste2', 'ResultadoTeste3', 'ResultadoTeste4'
]

var_dt = [
    'dataColetaTeste1', 'dataColetaTeste2', 'dataColetaTeste3', 'dataColetaTeste4'
]
```

```python
for name_old, name_new in rnm_columns.items():
    df20 = df20.withColumnRenamed(name_old, name_new)
    df21 = df21.withColumnRenamed(name_old, name_new)
    df22 = df22.withColumnRenamed(name_old, name_new)
```

```python
df20.select(*var_tests, *var_type_tests, *var_results, *var_dt).show()
```
