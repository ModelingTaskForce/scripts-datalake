```python
+-----------------------------------------+---------+
|vacina_nome                              |count    |
+-----------------------------------------+---------+
|COVID-19 ASTRAZENECA - ChAdOx1-S         |6988712  |
|COVID-19 SINOVAC/BUTANTAN - CORONAVAC    |82973933 |
|INF3                                     |27       |
|COVID-19 PFIZER - COMIRNATY              |101691303|
|COVID-19 ASTRAZENECA/FIOCRUZ - COVISHIELD|106924491|
|COVID-19 JANSSEN - Ad26.COV2.S           |4838676  |
+-----------------------------------------+---------+
```

```python
prop = []
for i in range(0,len(df1)):
    if df.nome_vac[i] == 'COVID-19 ASTRAZENECA - ChAdOx1-S': res = df.num_doses[i]/6988712
    elif df.nome_vac[i] == 'COVID-19 ASTRAZENECA/FIOCRUZ - COVISHIELD': res = df.num_doses[i]/106924491
    elif df.nome_vac[i] == 'COVID-19 JANSSEN - Ad26.COV2.S': res = df.num_doses[i]/4838676
    elif df.nome_vac[i] == 'COVID-19 PFIZER - COMIRNATY': res = df.num_doses[i]/101691303
    elif df.nome_vac[i] == 'COVID-19 SINOVAC/BUTANTAN - CORONAVAC': res = df.num_doses[i]/82973933
    elif df.nome_vac[i] == 'INF3': res = df.num_doses[i]/27
    prop.append(res)
```

```python
df1['prop_vac_with_dose'] = prop
```
